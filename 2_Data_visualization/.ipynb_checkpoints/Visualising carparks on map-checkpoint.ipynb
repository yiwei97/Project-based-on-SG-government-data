```python
# Read libraries and dataset

import folium
import pandas as pd
import numpy as np
from branca.element import Figure
import branca.colormap as cm
from folium.plugins import HeatMapWithTime
from folium.plugins import Geocoder
df = pd.read_csv("../1_Data_extraction/Output/carpark_combined.csv")
df = df.dropna()
df['lots_filled'] = df['total_lots'] - df['lots_available']
df['pct_filled'] = np.round(df['lots_filled'] / df['total_lots'] *100, 1)
df = df[df['total_lots']!=0]
df
```

```python
# Create bubble plot

variable = 'pct_filled'
m = folium.Map([1.35 ,103.8], zoom_start=10.5,width="%100",height="%100")
colormap = cm.StepColormap(colors = ['green','yellow','orange','red'],
                           index = [df[variable].describe()['min'],df[variable].describe()['25%'],df[variable].describe()['50%'],df[variable].describe()['75%'],df[variable].describe()['max']],
                           vmin = 0,
                           vmax = 100)
for loc, p in zip(zip(df["latitude"], df["longitude"]), df["pct_filled"]):
      folium.Circle(
      location=loc,
      radius=2,
      fill=True, 
      color=colormap(p)).add_to(m)
folium.plugins.Geocoder().add_to(m)
m
```

```python
# Create heat map
variable = 'lots_filled'
m = folium.Map([1.35 ,103.8], zoom_start=10.5,width="%100",height="%100")
heat_data = list(map(list, zip(df["latitude"],df["longitude"], df[variable])))
folium.plugins.HeatMap(heat_data).add_to(m)
folium.plugins.Geocoder().add_to(m)
m
```
